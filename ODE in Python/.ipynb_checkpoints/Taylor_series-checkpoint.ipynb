### Taylor Expansion of Function

Let's linearize $f(x) = sinx$ around $x = 0$
We will write the taylor expansion of $sinx$ around the origin.
Taylor series is used to express non polynomial function as a polynomial function, as polynomials are easier to deal with.

```python
import numpy as np
from matplotlib import pyplot as plt
import sympy as smp
import math
import time
```

```python
x0 = 0    # Point around which we find taylor expansion of the function
t = np.linspace((-np.pi + x0), (np.pi + x0), 100)
f = np.exp(t)
X = np.ones(len(t))*x0

# Y Axis
plt.plot(X, f)

# X Axis
plt.plot(t, np.zeros(len(t)))

# Exact plot
plt.plot(t, f, label="Exact")


# Taylor Approximation

x = smp.symbols("x")
G = smp.exp(x)

expr1 = G.subs(x,x0) + (smp.diff(G).subs(x, x0) * (x-x0))
expr2 = expr1 + (smp.diff(smp.diff(G)).subs(x, x0)*(x-x0)**2/math.factorial(2))
expr3 = expr2 + (smp.diff(smp.diff(smp.diff(G))).subs(x, x0)*(x-x0)**3/math.factorial(3))

F = np.zeros((len(t), 3))

for i in range(len(t)):
    F[i, 0] = expr1.subs(x, t[i])
    F[i, 1] = expr2.subs(x, t[i])
    F[i, 2] = expr3.subs(x, t[i])
    
plt.plot(t, F[:, 0], label = "$1^{st}$ Order Approximation")

plt.plot(t, F[:, 1], label = "$2^{nd}$ Order Approximation")

plt.plot(t, F[:, 2], label = "$3^{rd}$ Order Approximation")

plt.legend()
plt.show()
```
