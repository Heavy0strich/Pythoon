Plotting $f(x) = x^2+3x-5$

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
n = int(input("Enter the resolution: "))
h = 5/n                                        #step size
x = np.linspace(0, 5, n+1)
```

```python
f = x**2 +3*x-5
plt.plot(x, f, label = "x^2 +3x-5")
```

```python
# I want to interpolate, i.e., i want to find values at xp that is not in x

xp = eval(input("Enter a value: "))
k = int(xp/h)   # to find the nearest point in the array x

F = f[k] + ((f[k+1]-f[k])/(x[k+1]-x[k]))*(xp - x[k])

print(f"Approximation at f({xp}) = {F}")
print(f"Actual Value at {xp} is {xp**2 + 3*xp - 5}")
print(f"Error: {F - (xp**2 + 3*xp - 5)}")
```

Plotting Gaussian Function

$$f(x) = \frac{1}{\sigma\sqrt{2\pi}}e^{(\frac{-(x-\mu)^2}{2\sigma^2})};\space\space where \space x \space\epsilon\space[-5,5]$$

```python
n2 = int(input("Enter the resolution/no. of points"))
h = 10/n  #step size
x = np.linspace(-5,5,n+1)
```

```python
from math import sqrt, exp
# Parameters
sigma = 1
mu = 0

g = 1/(sigma*np.sqrt(2*np.pi))*np.exp(-(x - mu)**2/(2*sigma**2))
plt.plot(x, g, label = f"Gaussian with normal dis:{mu} and var:{sigma}")
plt.legend()
plt.title('Gaussian Distribution')

# math.sin or math.exp works on single number and np.sin or np.exp works on array. Since here we are plotting and 
# we are using vectors, thus use np.sin or np.exp in the function expression, so that you can compute it for each 
# element in x at the same time, rather then, passing each element of x and then storing it in the list
```

```python
import scipy.stats as stats
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

mean = 0; std = 1; variance = np.square(std)
x = np.arange(-5,5,.01)
f = np.exp(-np.square(x-mean)/2*variance)/(np.sqrt(2*np.pi*variance))

plt.plot(x,f)
plt.ylabel('gaussian distribution')
plt.show()
```
