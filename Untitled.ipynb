# Polynomial Data Fitting

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

First we generate data on the interval x $$ \Epsilon E [-1 to 1] from quadratic polynomial with additive noise

$y = 3*xj^2-2xj + \Alpha A


```python
#Generate and plot Data

m = 20 #generating points

np.random.seed(33) # seed random numbers generator
x = np.linspace(-1,1,m).T
y = 3*x**2 - 2*x + np.random.rand(m) #generating random y values for each x by adding some random noise as mentioned in the equation\
# The last term is the noise
plt.scatter(x,y);
```

## Polynomial Least square Fitting

```python
# Least Square fitting
n = 4 # we want to fit it to a degree 2 polynomial, thus n = 3

# Generate Vandermonde matrix of polynomials
vde = np.zeros((m,n))
for i in range(n):
    vde[:,i] = x**i
    
c = np.linalg.lstsq(vde,y,rcond=None)[0] #This function returns two objects and we require only the first one, thus index [0] used
fig, ax = plt.subplots()
ax.scatter(x,y) # plots the data
ax.plot(x,vde@c)

print(c) #Coefficients of the fitting polynomial
print(np.linalg.cond(vde)) #Gives Condition number, have to be greater than or equal to 1
```

# Polynomial Interpolation
Fitting a degreen m-1 polynomial (m coefficients) to m data points

$ p(x) = C_0 + C_1*x+....+C_m-1 * x^(m-1)

in contrast to least squares, we are gerring a degree m-1 polynomial to m points, so this is a lot more parameters to solve for and if you get more data points, you will be increasing the fegree of polynomial that you're fitting to it, implicitly. in a wat you're finding the polynomial that foes through all the data points, thus there are more parameters to fit. This is a much strixter risk restriction than the least squares fit which only minimizes the residual  

```python
#Polynomial interpolation

n = m

#generate Vandermonde matrix of polynomials

vde = np.zeros((m,n))
for i in range(n):
    vde[:,i] = x**i
    
c = np.linalg.lstsq(vde,y,rcond=None)[0]

# Evaluate the polynomial on the fine grid for plotting
xx = np.linspace(-1,1,100).T
p = np.zeros(xx.shape)
for i in range(n):
    p = p+c[i]*xx**i
    
fig,ax = plt.subplots()
ax.scatter(x,y)
#ax.plot(x,vde@c)
ax.plot(xx,p)
ax.set_ylim(-1,1)

print(np.linalg.cond(vde))
```

The polynomial if evaluted near to 1 or -1, you won't get a value between -1 and 1.
Thus fitting a reduced degree polynomial in the data gives a better fit, with least squares, we are solving for n degrees of freedom but in case of polynomial interpolation, we are solving for m degrees of freedom(m.n), thus stricter restrictions

The least square penalty acts as a regularization.
